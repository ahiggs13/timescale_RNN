```python
import yaml
import itertools
from pathlib import Path

kernel_taus = [0.1, 2]
tau_effects = ['all', 'decay']
distributions = ['groups', 'uniform']

for kernel_tau, tau_effect, distribution in itertools.product(kernel_taus, tau_effects, distributions):
    
    with open(f'base_{distribution}.yaml', 'r') as f:
        base = yaml.safe_load(f)
    
    config = base.copy()
    
    config['expt']['kernel_tau'] = kernel_tau
    config['model']['tau_effect'] = tau_effect
    config['model']['tau_distribution'] = distribution
    
    output_name = f"kerneltau{kernel_tau}_taueffect{tau_effect}_dist{distribution}.yaml"
    
    with open(output_name, 'w') as f:
        yaml.dump(config, f, default_flow_style=False)
    
    print(f"Generated: {output_name}")
```
